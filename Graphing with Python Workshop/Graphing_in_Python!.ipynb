<h1><center><span style="color:purple">A Background on Matplotlib</center></h1>
    
**Matplotlib is the most popular plotting module within Python. The module was originally developed by John Hunter, a PhD student studying neurobiology at the time. Originally Hunter was working across three different language, until he found Python. He found that Python was able to simplify much of his programming needs, but it still lagged behind Matlab in graphing and plotting. Hunter started developing the Matplotlib module to analyze human electrocorticography (ECoG) signals registered with 3D medical image data. From there it quickly grew to be the most popular plotting module in Python.**

<h1><center><span style="color:purple">A Background on seaborn</center></h1>

**seaborn is a plotting module for informative statistical graphs. It builds on top of Matplotlib and integrates closely with pandas data structures. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots.**

<h1><center><span style="color:purple">A Background on Plotly</center></h1>

**Plotly is a module that was created by a group of friends who originally worked in Silicon Valley. The group came up with a program, Plot.ly, that allowed multiple users to collaborate on graphs and charts that required large amounts of data. One of the unique features of the Plotly module is that the graphs and visualizations are interactive. Once the graphs are generated, a user can hover a bar chart or scatter plot to view the information on the graph.**

```python
#We will use pandas to store/work with our data. The data set is from NASA and contains information on various asteroids that the agency has identified

df = pd.read_csv('nasa.csv')

df
```

```python
#Let's get an idea of the types of data within the DataFrame

df.dtypes
```

```python
#Next we will check if there are any missing data points in the DataFrame

df.isnull().sum()
```

# The first characteristic we will plot is the how quickly the asteroids move per mile per hour. We will graph a histogram to look at the distribution. First we will take a look at the summary statistics of the column.

```python
df['Miles per hour'].describe()
```

```python
plt.figure('mph')
plt.hist(df['Miles per hour'], edgecolor = 'purple', lw = 2.5)
plt.axvline(df['Miles per hour'].median(), color='red', linestyle='dashed', linewidth=2, label = 'Median')
plt.axvline(df['Miles per hour'].mean(), color='orange', linestyle='dashed', linewidth=2, label = 'Median')
plt.title('Asteriods Speed in MPH')
plt.ylabel('Observations')
plt.xticks(np.arange(0, 110000, 10000), rotation = 'vertical')
plt.rcParams.update({'font.size': 16});
```

# The next characteristic we want to plot is whether or not the asteroid is hazardous. If we look at the data type above for the column, we can see that it is non-numeric, so we will have to convert it so that we can plot it in Matplotlib. Let's take a look at the elements within the Hazardous column.

```python
#We can see that all the elements within the column are boolean values, meaning True or False. Given this, we can convert the
#"False" values to 0 and "True" values to 1

df['Hazardous']
```

```python
df['Hazard Dummies'] = pd.get_dummies(df['Hazardous'], drop_first = True)

df
```

```python
hazard_array = df['Hazard Dummies'].value_counts()

hazard_array
```

```python
#Now that we have our dummies in an array, let's plot the information out into a barchart

plt.figure('hazard')
plt.bar(x = ['Not Hazardous', 'Hazardous'], height = hazard_array, color = ['lightblue', 'darkred'])
plt.title('Count of Hazardous Asteroids')
plt.ylabel('Count')
plt.yticks(np.arange(0, 5000, 500))
for i, v in enumerate(hazard_array):
    plt.text(y = v + 100, x = i -.15, s = str(round(v/sum(hazard_array),3)), fontweight='bold');
```

# We will now take a look at close approaches over time.

# A close approach is when an object (in this case an asteroid), travels closely to the Earth. Keep in mind that astronomers have a different definition of close, it is millions of miles instead of thousands or hundreds of miles.

#Close approaches can vary widely, the [Center for Near Earth Object Studies (CNEOS)](https://cneos.jpl.nasa.gov/ca/), lists objects with close aproaches from just under 2 million miles all the way out to over a billion miles!

```python
#First we will change the Close Approach column to the datetime data type

df['Close Approach Date'] = pd.to_datetime(df['Close Approach Date'])
```

```python
#px.scatter(data_frame = df, x = 'Close Approach Date', y = 'Miles per hour', size = 'Est Dia in Miles(max)', color = 'Hazard Dummies')

px.scatter(data_frame = df.loc[df['Hazardous'] == True], x = 'Close Approach Date', y = 'Miles per hour', size = 'Est Dia in Miles(max)', color = 'Orbit Uncertainity')
```

# Next we will look at the relationship between Eccentricity and Aphelion Distance.


# The Eccentricity of an asteroid's orbit describes the shape of an asteroid's movement. If the eccentricity is 0, then the orbit is a circle. If it is closer to 1, then the orbit is elliptical.


# The Aphelion Distance is the point at which the asteroid is farthest away from the sun. 


# Both of these features relate to the orbit of an asteroid, so we'll graph out a scatter plot of the two to see if there's a relationship



```python
px.scatter(data_frame = df, x = 'Eccentricity' ,y = 'Aphelion Dist', trendline = 'ols', trendline_color_override = "red")
```

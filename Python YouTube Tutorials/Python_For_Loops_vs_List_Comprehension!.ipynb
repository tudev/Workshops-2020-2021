```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

# **What are For Loops within Python?**



**The for statement in Python differs a bit from what you may be used to in C or Pascal. Rather than always iterating over an arithmetic progression of numbers (like in Pascal), or giving the user the ability to define both the iteration step and halting condition (as C), Python’s for statement iterates over the items of any sequence (a list or a string), in the order that they appear in the sequence.**

<sup>Source: [More Control Flow Tools](https://docs.python.org/3/tutorial/controlflow.html) from Python.org</sup>

# **What is List Comprehension within Python?**

**List comprehensions provide a concise way to create lists. Common applications are to make new lists where each element is the result of some operations applied to each member of another sequence or iterable, or to create a subsequence of those elements that satisfy a certain condition.**

<sup>Source: [Data Structures](https://docs.python.org/3/tutorial/datastructures.html) from Python.org</sup>

# **Creating a list of numbers**

## **For Loop**

```python
%%timeit
loop_list = []
for i in range(100000):
  loop_list.append(i)
```

## **List Comprehension**

```python
%%timeit
comp_list = [num for num in range(100000)]
```

# **Creating a list of numbers from a calculation**

## **For Loop**

```python
%%timeit
loop_list = []
for num in range(100000):
  loop_list.append(num**2)
```

## **List Comprehension**

```python
%%timeit
comp_list = [num**2 for num in range(100000)]
```

# **Conditional Statements**

## **For Loop**

```python
%%timeit
loop_list = []
for num in range(100000):
  if num%2 == 0:
    loop_list.append(num)
  #don't really need the else statement
  else:
    pass
```

## **List Comprehension**

```python
%%timeit
comp_list = [num for num in range(100000) if num%2 == 0]
```

# **Multiple Conditions**

## **For Loop**

```python
%%timeit
loop_list = []
for num in range(100000):
  if num%2 == 0:
    if num%5 == 0:
      loop_list.append(num)
```

## **List Comprehension**

```python
%%timeit
comp_list = [num for num in range(100000) if num%2 == 0 if num%5 == 0]
```

# **Nested List**

```python
while len(nested_list) < 100000:
  nested_list.append([random.randrange(1,25), random.randrange(1,25)]) 
```

```python
print(nested_list)
```

## **For Loop**

```python
%%timeit
loop_list = []
for i in nested_list:
  for j in i:
    loop_list.append(log(j))
```

## **List Comprehension**

```python
%%timeit
comp_list = [log(j) for i in nested_list for j in i]
```

#**Multiple Lists**

## **For Loop**

```python
%%timeit
x = []
y = []
for num in np.arange(-20,20,.01):
  x.append(num)
  y.append(num * np.sin(num))
```

## **List Comprehension**

```python
%%timeit
x = [num for num in np.arange(-20,20,.01)]
y = [num * np.sin(num) for num in x]
```

```python
plt.scatter(x,y);
```

# **Common NumPy Functions**

## **Concatenation**

**The concatenation function allows you to combine NumPy arrays together.**

```
fall_2020_grades = np.array([80, 85, 72, 73, 68, 90, 75, 95, 81, 77, 70, 64, 87, 71, 71, 75, 60, 74, 72])
summer_2020_grades = np.array([75, 70, 71, 80, 83, 84, 86, 67, 77, 70])
```

```
grades_2020 = np.concatenate((fall_2020_grades, summer_2020_grades))
print(grades_2020)
```

## **Summary Statistics: Average, Median, Quantile and Standard Deviation**
**The average function returns the mean of the array. The mean sums all the values in an object and divides it by the total observations within the object.**
<br/>
<br/>
**The median function returns the value in the middle of the sorted array. If the total observations within the object is odd, the median is calculated by taking the mean of the two middle values.**
<br/>
<br/>

**The quantile function returns the q-th quantile of the data along the specified axis.**
<br/>
<br/>

**The standard deviation function returns the variation or dispesion of the data within an object. The larger the standard deviation, the greater the variability within the data.**

```
print(f'The average grade for students in 2020 was: {np.average(grades_2020):.2f}\nThe median grade for students in 2020 was: {np.median(grades_2020)}')
print(f'The 75th quartile scored: {np.quantile(grades_2020, .75)}\nThe 25th quartile scored: {np.quantile(grades_2020, .25)}')
print(f'The standard deviation of the scores from 2020 was:{np.std(grades_2020):.2f}')
```

## **Max and Min**

```
print(f'The highest grade recorded for 2020 was: {np.max(grades_2020)}\nThe lowest grade recorded for 2020 was: {np.min(grades_2020)}')
```

## **Where**

**The where function returns the index position of elements in an array based on a given condition.**

```
for i in np.where(asteroid_array < 100):
  print(i)

small_asteroids = np.array([])

for i in np.where(asteroid_array < 100):
  small_asteroids = np.append(small_asteroids, asteroid_array[i])

print(small_asteroids)
```

## **Random**

**Is a package that produces pseudo random numbers using combinations of a BitGenerator to create sequences and a Generator to use those sequences to sample from different statistical distributions.**

**The NumPy random package is a great tool for data scientists that need to generate data sets to test their models. However the package should not be used for cryptographic because it generates pseudo random numbers. Instead Python users should consider the secrets package for cryptography.**

```
print(help(np.random))
```

```
random_array = np.random.random(size = 20)
print(random_array)

random_int_array = np.random.randint(low = 0, high = 100, size = 30)
print(random_int_array)

random_poisson_array = np.random.poisson(lam = 2, size = 25)
print(random_poisson_array)
```

## **Sort**
**The sort function modifies a NumPy array in place by ascending order by default. The sort function can also sort elements in descending order. Different sorting algorithms can be chosen by specifying the algorithm within the function.**

<br/>
<br/>

|Sorting Algorithm | Speed | Worst Case<img width=40/>    | Work Space | Stable |
|-|-|-| - | - |
|**quicksort**| $\large 1$ | $0(n^2)$  | $\large 0$| $\large No$ |
| **heapsort** | $\large 3$ | $0(n*log(n))$ | $\large 0$| $\large No$ |
| **mergesort** |$\large 2$ | $0(n*log(n))$ |$\Large \frac{-n}{2}$ | $\large Yes$|
| **timsort** |$\large 2$ |$0(n*log(n))$ |$\Large \frac{-n}{2}$ | $\large Yes$ |

<br/>

<sup>Source: [Documentation for NumPy, numpy.sort](https://numpy.org/doc/stable/reference/generated/numpy.sort.html)</sup>

```
random_int_array = -np.sort(-random_int_array)
print(random_int_array)
```

## **Trends of Specified Key Words**

**Interest Value: Numbers represent search interest relative to the highest point on the chart for the given region and time. A value of 100 is the peak popularity for the term. A value of 50 means that the term is half as popular. A score of 0 means there was not enough data for this term.**

<sup>Source: [pytrends introduction](https://github.com/GeneralMills/pytrends) on GitHub<sup/>

```python
kw_list = ['Python','JavaScript','C#']
```

```python
pytrend.categories()
```

```python
pytrend.build_payload(kw_list, cat=0, timeframe='all', geo='', gprop='')
country_df = pytrend.interest_by_region(resolution='COUNTRY', inc_low_vol=False, inc_geo_code=False).sort_values('Python',ascending=False)
country_df['name'] = country_df.index
country_df
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
merge=pd.merge(world,country_df,on='name')

merge.plot(column='Python',legend=True,cmap='viridis')
plt.title('Python Interest by Country');
```

## **Interest Over Time**

**Returns historical, indexed data for when the keyword was searched most as shown on Google Trends' Interest Over Time section**

<sup>Source: [pytrends introduction](https://github.com/GeneralMills/pytrends) on GitHub<sup/>

```python
program_df = pytrend.interest_over_time()
program_df.drop('isPartial',axis=1,inplace=True)
program_df
```

```python
program_df.plot()
plt.title('Interest in Programming Languages')
plt.ylabel('Interest over Time');
```

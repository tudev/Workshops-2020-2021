$y=2(x)+4$

```python
def func(x):
  return 2*x+4

interact(func, x=10);
```

$z=2(y)+2(x)+4$

```python
def multivariate_func(x,y):
  return 2*y + 2*x + 4

interact(multivariate_func,x=10,y=10);
```

# **Interactive Graphs**

```python
def linear_func(slope,intercept):
  plt.figure(2)
  x = np.linspace(-10,10,num=1000)
  plt.ylim(-10,10)
  plt.axvline(color='black')
  plt.axhline(color='black')
  plt.plot(x, slope * x + intercept, lw=5,color='red')
  plt.grid()
  plt.show()

interactive_plot = interactive(linear_func,slope=(-5.0,5.0),intercept=(-5.0,5.0))
output = interactive_plot.children[-1]
interactive_plot
```

<sup>Source: [Using Interact](https://ipywidgets.readthedocs.io/en/latest/examples/Using%20Interact.html) from ipywidgets documentation<sup/>

```python
def nonlinear_func(slope, intercept):
    plt.figure(2)
    x = np.linspace(-10, 10, num=1000)
    plt.ylim(-5, 10)
    plt.axvline(color='black')
    plt.axhline(color='black')
    plt.plot(x,x**(2) + slope * x + intercept, lw=5,color='blue')
    plt.grid()
    plt.show()

interactive_plot = interactive(nonlinear_func, slope=(-5.0, 5.0), intercept=(-5.0, 5.0))
output = interactive_plot.children[-1]
interactive_plot
```

**Utilized for data visualization. Facets Dive visualizes features, while Facets Overview returns descriptive statistics of the features.**

```python
!git clone https://github.com/pair-code/facets.git
!mv /content/facets/facets_overview/facets_overview/generic_feature_statistics_generator.py /content
```

```python
import generic_feature_statistics_generator
```

```python
hr_data = pd.read_csv('/content/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
hr_data.head()
```

```python
hr_data = hr_data.sample(frac=1)
hr_data.head()
```

```python
test_data = hr_data[:int(1470/3)]
train_data = hr_data[int(1470/3):]
```

```python
gfsg = generic_feature_statistics_generator.GenericFeatureStatisticsGenerator()
```

```python
proto = gfsg.ProtoFromDataFrames([{'name': 'train', 'table': train_data},
                                  {'name': 'test', 'table': test_data}])
protostr = base64.b64encode(proto.SerializeToString()).decode("utf-8")
```

```python
jsonstr_train = train_data.to_json(orient='records')
HTML_TEMPLATE = """
        <script src="https://cdnjs.cloudflare.com/ajax/libs/webcomponentsjs/1.3.3/webcomponents-lite.js"></script>
        <link rel="import" href="https://raw.githubusercontent.com/PAIR-code/facets/1.0.0/facets-dist/facets-jupyter.html">
        <facets-dive id="elem" height="600"></facets-dive>
        <script>
          var data = {jsonstr};
          document.querySelector("#elem").data = data;
        </script>"""
html = HTML_TEMPLATE.format(jsonstr=jsonstr_train)
display(HTML(html))
```

```python
jsonstr_test = test_data.to_json(orient='records')
HTML_TEMPLATE = """
        <script src="https://cdnjs.cloudflare.com/ajax/libs/webcomponentsjs/1.3.3/webcomponents-lite.js"></script>
        <link rel="import" href="https://raw.githubusercontent.com/PAIR-code/facets/1.0.0/facets-dist/facets-jupyter.html">
        <facets-dive id="elem" height="600"></facets-dive>
        <script>
          var data = {jsonstr};
          document.querySelector("#elem").data = data;
        </script>"""
html = HTML_TEMPLATE.format(jsonstr=jsonstr_test)
display(HTML(html))
```
